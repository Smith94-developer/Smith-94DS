```python
# According to a study, the daily average time spent by a user on a social media website is 50 minutes. To test the claim of this study, Ramesh, a researcher, takes a sample of 25 website users and finds out that the mean time spent by the sample users is 60 minutes and the sample standard deviation is 30 minutes. Based on this information, the null and the alternative hypotheses will be: Ho = The average time spent by the users is 50 minutes H1 = The average time spent by the users is not 50 minutes Use a 5% significance level to test this hypothesis.
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv(r"C:\Users\WELCOME\Downloads\Cell Phone Survey (1).csv.crdownload")


value_column = df['Value for the Dollar']
cell_phone_types = df['Type'].unique()


groups = []
for cell_phone_type in cell_phone_types:
    group = value_column[df['Type'] == cell_phone_type]
    groups.append(group)
    
f_statistic, p_value = stats.f_oneway(*groups)


print("F-statistic:", f_statistic)
print("p-value:", p_value)

plt.figure(figsize=(10, 6))
sns.kdeplot(value_column, shade=True, color='b')
plt.axvline(value_column.mean(), color='r', linestyle='dashed', linewidth=1)
plt.xlabel('Value for the dollar')
plt.ylabel('Density')
plt.title('Distribution of Value for dollar')
plt.legend(['Value for dollar', 'Mean'])
plt.show()
```

```python
#From the table given below, what is the mean of marks obtained by 20 students?
def calculate_weighted_mean(xi, fi):
    
    if len(xi) != len(fi):
        raise ValueError("The lengths of xi and fi must be the same.")

   
    weighted_sum = 0
    table_rows = []
    for x, f in zip(xi, fi):
        product = x * f
        weighted_sum += product
        table_rows.append([x, f, product])

   
    sum_frequencies = sum(fi)
  
    weighted_mean = weighted_sum / sum_frequencies

    print(" xi  |  fi  | xi * fi ")
    print("_____________________")
    for row in table_rows:
        print("{:3d}  | {:3d}  | {:5d}".format(row[0], row[1], row[2]))
    print("______________________")
    return weighted_mean

```

```python
xi = [3,4,5,6,7,8,9,10]
fi = [1,2,2,4,5,3,2,1]
mean_value = calculate_weighted_mean(xi, fi)
print("number of the  Mean:", mean_value)
```

```python
#For a certain type of computer, the length of time between charges of the battery is normally distributed with a mean of 50 hours and a standard deviation of 15 hours. John owns one of these computers and wants to know the probability that the length of time will be between 50 and 70 hours.
from scipy.stats import norm

mean = 50
std_dev = 15

z1 = (50 - mean) / std_dev
z2 = (70 - mean) / std_dev

probability = norm.cdf(z2) - norm.cdf(z1)
print("The probability of the length of time expected between 50 and 70 hours is:", probability)
```

```python
# Find the range of the following. g = [10, 23, 12, 21, 14, 17, 16, 11, 15, 19]
from statistics import mean, mode, median

def calculate_range(data):
    data_range = max(data) - min(data)
    return data_range

data = [10, 23, 12, 21, 14, 17, 16, 11, 15, 19]
data_mean = mean(data)

data_mode = mode(data)

data_median = median(data)

data_range = calculate_range(data)

print("Mean:", data_mean)
print("Mode:", data_mode)
print("Median:", data_median)
print("Range:", data_range)
```

```python
# It is estimated that 50% of emails are spam emails. Some software has been applied to filter these spam emails before they reach your inbox. A certain brand of software claims that it can detect 99% of spam emails, and the probability for a false positive (a non-spam 2 email detected as spam) is 5%. Now if an email is detected as spam, then what is the probability that it is in fact a non-spam email?
prior_prob_spam = 0.50
true_positive_prob = 0.99
false_positive_prob = 0.05

prob_non_spam_given_spam = (1 - false_positive_prob) * prior_prob_spam / ((1 - false_positive_prob) * prior_prob_spam + true_positive_prob * (1 - prior_prob_spam))

print("Probability of non-spam given detected as spam:", prob_non_spam_given_spam)
print("--------------------------------------------------------")
print("So, the probability that an email detected as spam is,")
print(" in fact, a non-spam email (false positive) is 0.95 or 95%.")
```

```python
# For a Binomial distribution, the number of trials(n) is 25, and the probability of success is 0.3. What’s the variability of the distribution?
import math 

n = 25
p = 0.3

variability = math.sqrt(n * p * (1 - p))

print("The variability of the distribution is:", variability)
```
