```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv(r'C:\Users\WELCOME\Downloads\honeyproduction 1998-2021.csv')
df
```

```python
#1 How has honey production yield changed from 1998 to 2021 ?

import pandas as pd 
import matplotlib.pyplot as plt

average_yield = df.groupby('year')['yieldpercol'].mean()
plt.figure(figsize=(12,6))
average_yield.plot(kind='line',marker='o')
plt.title('Honey Production Yield (1998-2021)')
plt.xlabel('year')
plt.ylabel('yieldpercol')
plt.grid(True)
plt.show()
```

```python
df.head()
df.info()
state_production = df.groupby('State')['totalprod'].sum().reset_index()
state_production = state_production.sort_values(by='totalprod', ascending=False)
plt.figure(figsize=(12, 6))
plt.bar(state_production['State'], state_production['totalprod'])
plt.xlabel('State')
plt.ylabel('Total Production')
plt.title('Honey Production by State')
plt.xticks(rotation=90)
plt.show()
```

```python
#2 Overtime, what are the major production trends across the states?
df.head()
df.info()
r2 = df.groupby('State')['totalprod'].sum().reset_index()
r2
r1 = r2.sort_values(by='totalprod', ascending=False)
r1
plt.figure(figsize=(18,12))
plt.barh(state_production['State'], state_production['totalprod'])

for i, totalprod in enumerate(r1['totalprod']):
plt.text(totalprod, i, str(totalprod), ha='left', va='center')

plt.xlabel('Total Production')
plt.ylabel('state')
plt.title('Honey Production by State')
plt.xticks(rotation=90)
plt.show()
```

```python
#3 Does the data show any trends in terms of the number of honey producing colonies and yield per colony before 2006, which was when concern over Colony Collapse Disorder spread nationwide?

df.head()
df.info()
```

```python
filtered_data = df[df['year'] < 2006]
filtered_data
```

```python
grouped_data = filtered_data.groupby('year').agg({'numcol': 'sum', 'yieldpercol': 'mean'}).reset_index()
grouped_data
```

```python
plt.figure(figsize=(10, 6))
plt.plot(grouped_data.index, grouped_data['numcol'], label='Number of Colonies')
plt.plot(grouped_data.index, grouped_data['yieldpercol'], label='Yield per Colony')
plt.xlabel('Year')
plt.ylabel('yieldpercol')
plt.title('Trends in Honey Production')
plt.legend()
plt.show()
```

```python
# Plotting the trends
plt.figure(figsize=(10, 6))
plt.bar(grouped_data['year'], grouped_data['numcol'])
plt.bar(grouped_data['year'], grouped_data['yieldpercol'])
plt.xlabel('Year')
plt.ylabel('yieldpercol')
plt.title('Trends in Honey Production')
 
plt.show()
```

```python
#4 Are there any patterns that can be observed between total honey production and value of production every year?
df['year'] = pd.to_datetime(df['year'], format='%Y')
```

```python
s1 = df.groupby(df['year']).sum().reset_index()
```

```python
plt.plot(s1['year'], s1['totalprod'], label='Total Honey Production')
plt.plot(s1['year'], s1['prodvalue'], label='Value of Production')
plt.xlabel('Year')
plt.ylabel('prodvalue')
plt.title('Total Honey Production vs. Value of Production')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(s1.index, s1['totalprod'], label='Total Honey Production',color='brown')
plt.bar(s1.index, s1['prodvalue'], label='Value of Production',color='green', alpha=0.5)
plt.xlabel('Year')
plt.ylabel('prodvalue')
plt.title('Total Honey Production vs. Value of Production')
plt.legend()
plt.show()
```
